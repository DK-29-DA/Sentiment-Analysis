```python
import csv
import string
import chardet
from google.colab import files
```

```python
# Upload the sentiment_lexicon.csv file (if not already uploaded)
uploaded_lexicon = files.upload()
```

```python
# Load sentiment lexicon from the uploaded CSV file
lexicon_file = 'sentiment_lexicon.csv'
sentiment_lexicon = {}
with open(lexicon_file, 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        word, sentiment = row
        sentiment_lexicon[word] = sentiment
```

```python
# Sentiment analysis function
def get_sentiment_score(words, sentiment_lexicon):
    positive_score = 0
    negative_score = 0
    negate = False

    for word in words:
        if word == "not" or word == "no":
            negate = not negate
            continue

        if word in sentiment_lexicon:
            sentiment = sentiment_lexicon[word]
            if negate:
                sentiment = 'negative' if sentiment == 'positive' else 'positive'
                negate = False
            if sentiment == 'positive':
                positive_score += 1
            elif sentiment == 'negative':
                negative_score += 1

    if positive_score > negative_score:
        return 'positive', positive_score, negative_score
    elif negative_score > positive_score:
        return 'negative', positive_score, negative_score
    else:
        return 'neutral', positive_score, negative_score
```

```python
# Analyze sentiment for each review text in the dataset
for index, row in data.head(20).iterrows():
    review_text = row['Summary']

    # Check if the value is a string, skip if it's not
    if isinstance(review_text, str):
        words = review_text.lower().translate(str.maketrans('', '', string.punctuation)).split()
        sentiment_info = get_sentiment_score(words, sentiment_lexicon)

        sentiment, positive_score, negative_score = sentiment_info
        print(f"Review: {review_text[:50]}...")  # Display only the first 50 characters of the review
        print(f"Words: {words}")
        print(f"Positive Score: {positive_score}")
        print(f"Negative Score: {negative_score}")
        print(f"Predicted Sentiment: {sentiment}")
        print()
    else:
        print("Skipping non-string value in the 'Review' column.")
```
